## People using at least basic sanitation services

```python
def filter_data(string):
    df_safe_water = df_data.loc[df_data['Indicator Name'].str.contains(string)]
    df_safe_water = df_safe_water[['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code',
       '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017']]

    country_1 = df_safe_water.loc[df_safe_water.isnull().sum(axis = 1) > 5]['Country Code'].unique()

    df_safe_water = df_safe_water.loc[~df_safe_water['Country Code'].isin(country_1)]
    df_safe_water = df_safe_water.drop(['Country Code', 'Indicator Code'], axis=1)
    df_safe_water['Indicator Name'] = df_safe_water['Indicator Name'].str.replace(string + ' \(% of population\)', 'overall')
    df_safe_water['Indicator Name'] = df_safe_water['Indicator Name'].str.replace(string + ', rural \(% of rural population\)', 'rural')
    df_safe_water['Indicator Name'] = df_safe_water['Indicator Name'].str.replace(string + ', urban \(% of urban population\)', 'urban')
    df_safe_water = df_safe_water.set_index(['Country Name', 'Indicator Name'], append=False)
    df_safe_water.columns.name = 'YEAR'
    df_safe_water = df_safe_water.unstack('Indicator Name')
    df_safe_water = df_safe_water.stack('YEAR')
    df_safe_water.to_csv(string + '.csv', index=True, header=True)
```

```python
filter_data('People using at least basic sanitation services')
```
